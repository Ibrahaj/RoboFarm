```python
from threading import Thread, Lock
import logging
import time
```

```python

logging.basicConfig(level=logging.DEBUG,
                    format='(%(threadName)-10s) %(message)s',
                    )
```

```python
class Counter(Thread):
          
    def __init__(self, name=str, value=0):
        
        super().__init__(group=None, target=None, 
                              name=name)
        self.value = value
        self.lock = Lock()

    def run(self):
            
            time.sleep(0.5)
            logging.debug('Waiting for lock')
            with self.lock:
                self.increment()
            logging.debug('lock released')
            return
    def increment(self):
        logging.debug('incrementing')
        self.value = self.value + 1
        logging.debug(f'incremet val = {self.value}')
```

```python
check = True
val = 0
oldval = None 
while check:
    logging.debug(f'start val = {val}')
    count = Counter('one',val)
    count.start()
    logging.debug(f'count val = {count.value}')    
    logging.debug(f'val = {val}')
    if val<10:
        logging.info('moving')
        if oldval is not None and val==oldval:
            
            logging.info(f'value = old value = {val}')
        else:
            oldval = val
            
        val = count.value
        logging.info(f'value updated = {val}')
        logging.debug('sleeping for 0.25 sec')      
        time.sleep(0.25)
    else:
        logging.info('goal is reached')
        check = False
```
